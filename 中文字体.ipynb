```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['font.family'] = ['SimHei'] #或者 ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('正弦曲线')
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.show()
```
