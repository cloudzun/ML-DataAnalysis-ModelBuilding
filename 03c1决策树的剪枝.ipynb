# 数据预处理

## 导入数据

````python
import pandas as pd #导入Pandas
df_sales = pd.read_csv('易速鲜花订单记录.csv') #载入数据
df_sales.head() #显示头几行数据 
````

## 整理日期格式

````python
df_sales = pd.read_csv('易速鲜花订单记录.csv') #载入数据
print('日期范围（格式转化前）: %s ~ %s' % (df_sales['消费日期'].min(), df_sales['消费日期'].max())) #显示日期范围（格式转换前）
df_sales['消费日期'] = pd.to_datetime(df_sales['消费日期']) #转换日期格式
print('日期范围（格式转化后）: %s ~ %s' % (df_sales['消费日期'].min(), df_sales['消费日期'].max()))#显示日期范围
````

````python
df_sales = df_sales.loc[df_sales['消费日期'] < '2021-06-01'] #只保留整月数据
print('日期范围（删除不完整的月份）: %s ~ %s' % (df_sales['消费日期'].min(), df_sales['消费日期'].max())) #显示日期范围
````

## 数据可视化

## 数据清洗

````python
df_sales = df_sales.drop_duplicates() #删除重复的数据行
````

## 特征工程

````python
df_sales['总价'] = df_sales['数量'] * df_sales['单价'] #计算每单的总价
df_sales.head() #显示头几行数据  
````

### 构建User用户表

````python
df_sales_3m = df_sales[(df_sales.消费日期 > '2020-06-01') & (df_sales.消费日期 <= '2020-08-30')] #构建仅含头三个月数据的数据集
df_sales_3m.reset_index(drop=True) #重置索引
````

````python
df_user = pd.DataFrame(df_sales['用户码'].unique()) #生成以用户码为主键的结构
df_user.columns = ['用户码'] #设定字段名
df_user.head() #显示头几行数据
df_R_value = df_sales_3m.groupby('用户码').消费日期.max().reset_index() #找到每个用户的最近消费日期，构建df_R_value对象
df_R_value.columns = ['用户码','最近购买日期'] #设定字段名
df_R_value['R值'] = (df_R_value['最近购买日期'].max() - df_R_value['最近购买日期']).dt.days #计算最新日期与上次消费日期的天数
df_user = pd.merge(df_user, df_R_value[['用户码','R值']], on='用户码') #把上次消费距最新日期的天数（R值）合并至df_user结构
df_F_value = df_sales_3m.groupby('用户码').消费日期.count().reset_index() #计算每个用户消费次数，构建df_F_value对象
df_F_value.columns = ['用户码','F值'] #设定字段名
df_user = pd.merge(df_user, df_F_value[['用户码','F值']], on='用户码') #把消费频率(F值)整合至df_user结构
df_M_value = df_sales_3m.groupby('用户码').总价.sum().reset_index() #计算每个用户三个月消费总额，构建df_M_value对象
df_M_value.columns = ['用户码','M值'] #设定字段名
df_user = pd.merge(df_user, df_M_value, on='用户码') #把消费总额整合至df_user结构
df_user #显示用户表结构
````

## 构建标签

````python
df_user_1y = df_sales.groupby('用户码')['总价'].sum().reset_index() #计算每个用户整年消费总额，构建df_user_1y对象
df_user_1y.columns = ['用户码','年度LTV'] #设定字段名
df_user_1y.head() #显示头几行数据
df_LTV = pd.merge(df_user, df_user_1y, on='用户码', how='left') #构建整体LTV训练数据集
df_LTV #显示df_LTV
````

## 创建特征集和标签集

````python
X = df_LTV.drop(['用户码','年度LTV'],axis=1) #特征集
X.head() #显示特征集
````

````python
y = df_LTV['年度LTV'] #标签集
y.head() #显示标签集
````

## 拆分训练集、验证集和测试集

````python
from sklearn.model_selection import train_test_split
# 先拆分训练集和其它集
X_train, X_rem, y_train, y_rem = train_test_split(X,y, train_size=0.7,random_state = 0)
# 再把其它集拆分成验证集和测试集 
X_valid, X_test, y_valid, y_test = train_test_split(X_rem,y_rem, test_size=0.5,random_state = 0)
````

````python
from sklearn.preprocessing import MinMaxScaler #导入归一化缩放器
scaler = MinMaxScaler() #创建归一化缩放器
X_train = scaler.fit_transform(X_train) #拟合并转换训练集数据
X_valid = scaler.transform(X_valid) #转换验证集数据
X_test = scaler.transform(X_test) #转换测试集数据
````

这段代码使用scikit-learn库的`MinMaxScaler`工具对数据进行归一化缩放。归一化是一种将数值特征重新缩放到特定范围（通常为0到1）的数据预处理技术。归一化有助于加快算法收敛速度，避免某些变量很大或很小导致的影响过大或过小，并使模型可以在不同尺度的特征间平衡权重。这里是代码的详细解释：

1. 从`sklearn.preprocessing`模块导入`MinMaxScaler`工具。
   ```python
   from sklearn.preprocessing import MinMaxScaler
   ```

2. 创建一个新的`MinMaxScaler`实例，命名为`scaler`。
   ```python
   scaler = MinMaxScaler()
   ```

3. 使用`fit_transform()`函数，让`scaler`先学习`X_train`(训练集数据)中特征的最大值和最小值，然后根据这些值将其`X_train`归一化。归一化后的训练数据再赋值给`X_train`。
   ```python
   X_train = scaler.fit_transform(X_train)
   ```

4. 使用`transform()`函数，根据训练数据集的最大值和最小值将验证集数据`X_valid`和测试集数据`X_test`进行归一化。
   ```python
   X_valid = scaler.transform(X_valid)
   X_test = scaler.transform(X_test)
   ```

完成这些步骤后，训练集、验证集和测试集的特征都已经进行归一化处理。现在可以使用处理后的数据训练模型，并在验证集和测试集上检验模型性能。需要注意的是，只需对训练集调用`fit_transform`方法，而对验证集和测试集只调用`transform`方法，确保归一化处理是基于训练集的统计信息，避免数据泄漏。

# 训练模型，决策树剪枝

通过剪枝，我们可以控制决策树生成时的深度，避免过于精准地切分训练集数据点，从而导致模型无法泛化到其他数据集。

````python
from sklearn.tree import DecisionTreeRegressor #导入决策树回归模型
model_dtr = DecisionTreeRegressor() #创建决策树回归模型
model_dtr_cut = DecisionTreeRegressor(max_depth=3) #创建深度为3的决策树回归模型
````

这段代码导入了决策树回归模型（DecisionTreeRegressor），并创建了两个不同的决策树回归实例。以下解释这两个实例的含义以及它们在代码中的实现方法。

1. 决策树回归：决策树回归是一种回归模型，其主要目标是将数据集递归分割成一定数目的子集，使得每个子集中的目标变量尽可能地保持相似的取值范围。在预测时，给定输入特征，模型将沿着从根节点到叶节点的路径进行决策，并根据叶子节点上目标变量的平均值做出预测。

   创建决策树回归模型：
   ```python
   model_dtr = DecisionTreeRegressor()
   ```

2. 决策树回归（深度限制）：决策树回归的一个问题是容易发生过拟合现象，特别是在树的深度较大时。为了解决这个问题，可以限制决策树的最大深度。此处，创建了一个最大深度为3的决策树回归模型。这意味着模型会在最大深度为3的条件下进行分裂，从而降低过拟合风险。

   创建深度为3的决策树回归模型：
   ```python
   model_dtr_cut = DecisionTreeRegressor(max_depth=3)
   ```

现在，已经创建了两个决策树回归模型：一个没有深度限制，另一个深度限制为3。接下来可以使用训练数据训练这些模型，并在验证集和测试集上对它们的性能进行评估。

````python
model_dtr.fit(X_train, y_train) #拟合决策树模型
model_dtr_cut.fit(X_train, y_train) #拟合深度为3的决策树模型
y_valid_preds_dtr = model_dtr.predict(X_valid) #用决策树模型预测验证集
y_valid_preds_dtr_cut = model_dtr_cut.predict(X_valid) #用深度为3的决策树模型预测验证集
````

这段代码训练了两个决策树回归模型（没有深度限制的决策树回归模型和深度限制为3的决策树回归模型），然后使用这两个模型分别对验证集进行预测。以下详细解释代码的作用和步骤。

1. 拟合决策树模型：这一步使用训练集（X_train和y_train）对没有深度限制的决策树回归模型进行训练。

   ```python
   model_dtr.fit(X_train, y_train)
   ```

2. 拟合深度为3的决策树模型：这一步使用训练集（X_train和y_train）对深度限制为3的决策树回归模型进行训练。

   ```python
   model_dtr_cut.fit(X_train, y_train)
   ```

3. 用决策树模型预测验证集：这一步使用没有深度限制的决策树回归模型对验证集（X_valid）进行预测，并将预测结果存储在y_valid_preds_dtr中。

   ```python
   y_valid_preds_dtr = model_dtr.predict(X_valid)
   ```

4. 用深度为3的决策树模型预测验证集：这一步使用深度限制为3的决策树回归模型对验证集（X_valid）进行预测，并将预测结果存储在y_valid_preds_dtr_cut中。

   ```python
   y_valid_preds_dtr_cut = model_dtr_cut.predict(X_valid)
   ```

现在，已经使用两个决策树回归模型对验证集进行了预测。接下来可以计算这些模型在验证集上的性能指标，例如R²分数、均方误差等。这将有助于比较这两个模型的表现以及它们与其他回归模型（线性回归、Lasso回归和Ridge回归）的性能差异。

## 评估模型

````python
from sklearn.metrics import r2_score,   median_absolute_error #导入Sklearn评估模块
print('训练集上的R平方分数-决策树: %0.4f' % r2_score(y_train, model_dtr.predict(X_train)))
print('训练集上的R平方分数-深度为3的决策树: %0.4f' % r2_score(y_train, model_dtr_cut.predict(X_train)))
print('测试集上的R平方分数-决策树: %0.4f' % r2_score(y_valid, model_dtr.predict(X_valid)))
print('测试集上的R平方分数-深度为3的决策树: %0.4f' % r2_score(y_valid, model_dtr_cut.predict(X_valid)))
````

这段代码计算了两个决策树回归模型（无深度限制的决策树和深度限制为3的决策树）在训练集和验证集上的R²分数。接下来，我将分析输出的结果：

1. 训练集上的R²分数 - 决策树：1.0000
   这个分数表示没有深度限制的决策树模型在训练集上的表现是完美的。模型能完全捕获训练数据中目标变量的变化。然而，在这种情况下，模型可能发生过拟合，因为它在训练数据上表现得太好了。

2. 训练集上的R²分数 - 深度为3的决策树：0.8577
   这个分数表示深度限制为3的决策树模型在训练集上的拟合程度较好。这意味着模型捕获了训练数据中目标变量的大部分变化，但可能没有过拟合。

3. 测试集上的R²分数 - 决策树：0.2694
   这个分数表示没有深度限制的决策树模型在验证集上的表现相对较差。它只解释了目标变量大约27%的方差。这进一步证实了模型可能过拟合训练数据，并在验证数据（即新数据）上泛化性能较差。

4. 测试集上的R²分数 - 深度为3的决策树：0.3284
   这个分数表示深度限制为3的决策树模型在验证集上的表现要优于没有深度限制的模型。该模型解释了目标变量约33%的方差。虽然这个分数仍然不高，但它表明简化模型结构（通过限制深度）可能有助于改善在新数据上的泛化能力。

总之，深度限制为3的决策树模型在训练集和验证集上的表现相对较好，更能泛化到新数据。同时，没有深度限制的决策树模型可能过拟合，导致在验证集上的性能下降。尽管深度限制为3的模型在验证集上的R²分数仍有改进的空间，但就当前情况而言，它是两者中更优的选择。
